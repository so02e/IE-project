```python
import pandas as pd
```

```python
data = pd.read_csv("data/전국CCTV표준데이터.csv", encoding='ansi')
data
```

```python
cctv_suwon = data[data['소재지도로명주소'].str.contains('수원시',na = False)]
```

```python
cctv_suwon
```

```python
cctv_suwon.to_csv("output/수원_cctv_정보")
```

```python
import folium
import json

geo_path = 'data/31.수원시_행정경계(읍면동).geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))

suwon_map = folium.Map(location=[37.2752207,127.0138527], zoom_start=13, tiles="cartodbpositron")


    

fmap=folium.Choropleth(geo_data = geo_str,
                       color = "black",
                       nan_fill_color='black',
                       fill_opacity=0.1,
                       fill_color = None).add_to(suwon_map)




for name, lat, lng in zip(cctv_suwon.제공기관명, cctv_suwon.위도, cctv_suwon.경도):
    folium.CircleMarker([lat, lng],
                        radius=9,         # 원의 반지름
                        fill=True,
                        fill_color='blue',    # 원을 채우는 색
                        fill_opacity=0.7, # 투명도    
                        popup=name
    ).add_to(suwon_map)

    

display(suwon_map)

suwon_map.save('output/suwon_map_CCTV.html')
```
