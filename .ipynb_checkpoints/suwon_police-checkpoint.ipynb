```python
import pandas as pd
```

```python
data = pd.read_csv("data/suwon_police_data.csv")
data
```

```python
import folium
import json

geo_path = 'data/suwon_police.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))

suwon_map = folium.Map(location=[37.2752207,127.0138527], zoom_start=13, tiles="cartodbpositron")


    

fmap=folium.Choropleth(geo_data = geo_str,
                       color = "black",
                       nan_fill_color='black',
                       fill_opacity=0.1,
                       fill_color = None).add_to(suwon_map)




for name, lat, lng in zip(data.관서명, data.WGS84위도, data.WGS84경도):
    folium.CircleMarker([lat, lng],
                        radius=6,         # 원의 반지름
                        fill=True,
                        fill_color='red',    # 원을 채우는 색
                        fill_opacity=0.3, # 투명도    
                        popup=name
    ).add_to(suwon_map)

    

display(suwon_map)

suwon_map.save('output/suwon_map_police.html')
```
